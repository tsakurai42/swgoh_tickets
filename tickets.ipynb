```python
try:
    from PIL import Image
except ImportError:
    import Image
import pytesseract
from PIL import ImageOps
import pandas as pd
import os
```

```python
dict = {}
namebox = (606,292,800,676)
ticketbox = (1100,292,1225,682)
with os.scandir('tickets/lifetime') as files:
    for each_file in files:
        im = Image.open(f'tickets/lifetime/{each_file.name}')
        region = im.crop(namebox)
        regioninv = ImageOps.invert(region)
        namestring = pytesseract.image_to_string(regioninv)
        name_arr = namestring.split('\n')
        list_of_names = []
        for i in range(0,len(name_arr),3):
            list_of_names.append(name_arr[i])
            
        region2 = im.crop(ticketbox)
        regioninv2 = ImageOps.invert(region2)
        ticketstring = pytesseract.image_to_string(regioninv2)
        ticket_arr = ticketstring.split('\n')
        order = 0
        for i in range(len(ticket_arr)):
            if ticket_arr[i].isnumeric():
                dict[list_of_names[order]]={'lifetime':ticket_arr[i]}
                order += 1
with os.scandir('tickets/daily') as files:
    for each_file in files:
        im = Image.open(f'tickets/daily/{each_file.name}')
        region = im.crop(namebox)
        regioninv = ImageOps.invert(region)
        namestring = pytesseract.image_to_string(regioninv)
        name_arr = namestring.split('\n')
        list_of_names = []
        for i in range(0,len(name_arr),3):
            list_of_names.append(name_arr[i])
            
        region2 = im.crop(ticketbox)
        regioninv2 = ImageOps.invert(region2)
        ticketstring = pytesseract.image_to_string(regioninv2)
        ticket_arr = ticketstring.split('\n')
        order = 0
        for i in range(len(ticket_arr)):
            if ticket_arr[i].isnumeric():
                dict[list_of_names[order]]['daily']=ticket_arr[i]
                order += 1
print(dict)
```

```python
testpd = pd.DataFrame.from_dict(dict,orient='index')
testpd = testpd.iloc[testpd.index.str.lower().argsort()]
```

```python
testpd.to_csv('tix.csv')
```
